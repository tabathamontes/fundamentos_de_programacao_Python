FORMAS DE PORCENTAGEM

1. Calcular uma porcentagem de um valor
Para calcular uma porcentagem de um valor, você pode usar a fórmula:


valor = 100
percentual = 20
resultado = (valor * percentual) / 100
print(resultado)  # Saída: 20.0


2. Acrescentar uma porcentagem a um valor
Para acrescentar uma porcentagem a um valor, você pode usar a fórmula:

valor = 100
percentual = 20
resultado = valor + (valor * percentual / 100)
print(resultado)  # Saída: 120.0

3. Subtrair uma porcentagem de um valor
Para subtrair uma porcentagem de um valor, você pode usar a fórmula:

valor = 100
percentual = 20
resultado = valor - (valor * percentual / 100)
print(resultado)  # Saída: 80.0


4. Descobrir a porcentagem entre dois valores
Para descobrir a porcentagem que um valor representa de outro, você pode usar a fórmula:

parte = 20
todo = 100
percentual = (parte / todo) * 100
print(percentual)  # Saída: 20.0


IMPORTANTE!!!
Se o percentual estiver na forma inteira, você deve dividi-lo por 100 para transformá-lo em decimal. Por exemplo, 25% se torna 0,25 quando dividido por 100.

Se o percentual já estiver em decimal, como 0,25, não é necessário fazer mais nada. Ele já está na forma decimal correta


EX:
CONVERSÃO
percentual = 20  # Percentual em forma inteira

novo_salario = salario + (salario * percentual / 100)  # Dividindo por 100 para converter a forma inteira para decimal

SEM CONVERSÃO
percentual = 0.15  # Percentual em forma decimal
novo_salario = salario + (salario * percentual)



Essa fórmula é usada para calcular um novo salário após um aumento percentual.
SE EU FIZER ASSIM:
novo_salario = salario * (1 + percentual)

salario: o salário atual.

percentual: a taxa de aumento expressa como um decimal (por exemplo, 10% seria 0,10).

(1 + percentual): Adiciona 1 ao percentual para representar o valor total do salário após o aumento. Por exemplo, se o percentual for 0,10 (10%), então ( 1 + 0,10 = 1,10 ).

salario * (1 + percentual): Multiplica o salário atual pelo valor total calculado no passo anterior. Isso resulta no novo salário, que inclui o aumento


Para descobrir quantos porcento um valor representa em relação a outro, você pode seguir estes passos simples:

Divida o valor desejado pelo valor total.
Multiplique o resultado por 100.

A fórmula é:
porcentagem=(valor_desejado / valor_total​) × 100
Por exemplo, se você quer saber quantos porcento 50 é de 200:
Porcentagem=(50 /200​) × 100 = 25%
